## 1. Introduction

Minor irrigation ponds are man-made banked earth structures designed to capture rainwater that constitute a major source of irrigation water in South Indian villages. De-silting of these ponds, usually done in the dry season, is necessary to keep up the water storage capacity and has the added benefit of providing a source of nutrient rich silt, which can be applied to agricultural land to improve soil fertility.

This project aims to guide the operational execution of silt application on farmer land. It will be executed by hiring fleets of tractors and providing silt to farmers. Selecting villages can be done over many criteria. The tractors can target villages based on: maximum current demand for silt, maximum opportunity for increase in productivity, villages with the maximum agricultural land, villages spatially closest to other villages etc.

Agent Based Modelling is a good way to simulate the problem because of the individual decisions that the many actors are taking. There are two key types of actors - villages that decide how much silt they want and tractor fleets that go around delivering the silt. Secondly, in reality decisions will be made on a local spatial scale, due to on the ground management realities (supervisors have local knowledge) - instead of a global optimization. Agent based modelling is a good approximation of this reality. 

This model looks at two ways of optimizing operation - going to where current demand is maximum and going to where the silt quality is the highest. Success is measured by the overall pace at which silt is applied onto farmer land (benefit to the enterprise) and the change in total productivity (benefit to the farmers).

## 2. Methods

This paper uses a grid based spatial agent based model.  

### 2.1 Space

The state of Telangana has about 10,000 villages spread across 25 million acres of land. Farmland is spread around these villages. 60% of India's total land is in farming - for Telangana this comes to 15 million acres or about 1500 acres per village, which is similar to anecdotal accounts by villagers

To capture these on the ground charachteristics in a simplified manner the state is modelled by a 10 x 10 grid. This gives a total area of 100 units. I want 60% of this land to be in farms and so 60 "villages" are placed in this space. 

There are two spaces similar spaces, one to place the villages and one for the tractors to move in. 

### 2.2 Agents

#### 2.2.1 Villages

Each village is placed at a permanent random location at the start of the simulation. The people living in the villages are farmers and own the land surrounding the village. For simplicity, I will assume that the village representation in the model is both the land and the farmers. 

Due to the varying quality of soil, water availability, access to information, seed choice, spending on fertilizers and pesticides the villages have different productivity of cotton ranging from 400 kg per acre to 800 kg per acre. Each village will be randomly assigned a base productivity at the start of the simulation. Each village will also be assigned an amount of land that they farm. Each village also has an irrigation tank that they use for watering their crops. Each tank has a different quality of silt, which depends on the upstream catchement - and not the local soil. This quality will be assigned randonly as well (although there is spatial relation to other villages).  


After the Green Revolution, farmers moved from using the silt - the only source of fertilizer at that time to artificial fertilizers. This led to a huge increase in productiviy coupled with a large increase in input costs with the new equilibrium being the more artificial input intensive way of farming. Although it was used centuries ago, the use of silt has been disontinued and even if a village stands to benefit from application - farmers will only demand it if they have seen some benefit in the recent past or have heard about it. Each village will be given some early adopters who will demand silt. Applying silt will increase the productivity of their land, based on the quality of the land - therebey increasing demand in this and nearby villages.

In summary village parameters consist of:

Constant parameters 
1. base_yield: Base cotton yield (in kg per acre)
2. cotton_acres: Acres under cotton production (in acres)
3. silt_quality: Quality of silt in irrigation tank (a percentage increase from baseline yield ranging from 0.0 - 1.0)

Variable Parameters
4. silt_acres: Acres on which silt has been applied
5. inc_prod: Increase in cotton productivity on treatment acres
6. silt_demand: Silt demanded (random initial number - grows due to increase in productivity)

#### 2.2.2 Tractor Fleets

The supervisor deals with hiring tractors and can hire as many tractors as required to meet the local demand at that time step. There are no moving costs, only in terms of time. 

Tractors want to maximimze the payments that they get during the simulation. They recieve payments upfront for each acre that they apply silt. They also recieve additional performance based payments depending on the increase in productivity. However, these are calculated seperately as gains made by the farmer. Initially this will just depend on the change in productivity though later some stochasticity may be introduced to simulate floods or droughts.  

### 2.3 Model

#### 2.3.1 Model Step

There are 3 months of dry season during which tractors can transport silt from the tanks to nearby farms. Each time step is representative of the three month dry season. At the end of the dry season the villagers will grow cotton and benefit from the cumulative silt applied during the previous. At this point the new silt demand will be calculated. The tractors can now move to a new place of their choosing.

Silt demand by a village is critical - and the determining factor of how much is delivered. It is calculated as follows and should replicate an S curve adoption:


$$\text{Silt Demand by a Village} =\left( \text{base demand percent} + \text{Silt Quality} \times \frac{\text{Silt Applied}}{\text{Total Acreage in Village}} \right)\times( \text{total acreage} - \text{Silt Applied})$$

Currently the tractor fleet will supply all the silt demanded by a village.

Some hand calculated demand values to test adoption rates. These will vary based on silt quality, total acreage and base demand rates.

$$\text{Silt Demand by a Village} =\left( 0.05 + 0.5 \times \frac{100}{1000} \right) \times( 1000 - 100) = 90 \text{ acres}$$

$$\text{Silt Demand by a Village} =\left( 0.05 + 0.5 \times \frac{190}{1000} \right) \times( 1000 - 190) = 117.45 \text{ acres}$$

$$\text{Silt Demand by a Village} =\left( 0.05 + 0.5 \times \frac{307.45}{1000} \right) \times( 1000 - 307.45) = 141.09 \text{ acres}$$

### Model Class

```python
class Model(object):
    
    def __init__(self, grid_size, num_villages, num_tractors):
    
        #set model Parameters
        self.grid_size = grid_size
        self.num_villages = num_villages
        self.num_tractors = num_tractors
    
        #set state variables
        self.t = 0
        self.space = numpy.array((0,0))
        self.tractorspace = numpy.array((0,0))
        self.villages = []
        self.tractors = []
        self.num_applied = 0
        self.total_applied = 0 #used to track
        
        #set history variables
        self.history_space = []
        self.history_tractorspace = []
        self.history_acres_applied = []        
        self.history_total_acres_applied = []        

        #call setup methods to initialize things
        self.setup_space()
        self.setup_tractorspace()
        self.setup_villages()
        self.setup_tractors()
    
    def setup_space(self):
        """
        Method to setup our space.
        """
        # Initialize a space with a NaN's
        self.space = numpy.full((self.grid_size, self.grid_size), numpy.nan)

    def setup_tractorspace(self):
        """
        Method to setup our tractor space.
        """
        # Initialize a space with a NaN's
        self.tractorspace = numpy.full((self.grid_size, self.grid_size), numpy.nan)

    def setup_villages(self):
        """
        Method to setup our space.
        """
        # First, begin by creating all agents without placing them.
        for i in xrange(self.num_villages):
            self.villages.append(Village(model=self,
                                          village_id=i,
                                          silt_applied=False,
                                          cotton_acres = numpy.random.randint(1000,2000), 
                                          base_yield = numpy.random.randint(4,6), 
                                          silt_quality = numpy.random.uniform(0,1), 
                                          silt_acres = numpy.random.randint(1,10), 
                                          inc_prod = 1, 
                                          base_silt_demand = numpy.random.uniform(0.0,0.05),  
                                          total_silt_demand = 10))

            
        # Second, once created, place them into the space.
        for village in self.villages:
            # Loop until unique
            is_occupied = True
            while is_occupied:
                # Sample location
                random_x = numpy.random.randint(0, self.grid_size)
                random_y = numpy.random.randint(0, self.grid_size)
                
                # Check if unique
                if numpy.isnan(self.space[random_x, random_y]):
                    is_occupied = False
                else:
                    is_occupied = True
            
            # Now place the person there by setting their ID.
            self.space[random_x, random_y] = village.village_id
     
     
    def setup_tractors(self):
        # First, begin by creating all agents without placing them.
        for i in xrange(self.num_tractors):
            self.tractors.append(Tractor(model=self,
                                      tractor_id=i, acres_applied = 0, profits = 0))
                                    
        
        # Second, once created, place them into the space.
        for tractor in self.tractors:
            # Loop until unique
            is_occupied = True
            while is_occupied:
                # Sample location
                random_x = numpy.random.randint(0, self.grid_size)
                random_y = numpy.random.randint(0, self.grid_size)
                
                # Check if unique
                if numpy.isnan(self.tractorspace[random_x, random_y]):
                    is_occupied = False
                else:
                    is_occupied = True
            
            # Now place the person there by setting their ID.
            self.tractorspace[random_x, random_y] = tractor.tractor_id
    
    def get_neighborhood(self, x, y, distance=1):
        """
        Get a Moore neighborhood of distance from (x, y).
        """
        neighbor_pos = [ ( x % self.grid_size, y % self.grid_size)
                                for x, y in itertools.product(xrange(x-distance, x+distance+1),
                                xrange(y-distance, y+distance+1))]
        return neighbor_pos
    
    def get_neighbors(self, x, y, distance=1):
        """
        Get any neighboring persons within distance from (x, y).
        """
        neighbor_pos = self.get_neighborhood(x, y, distance)
        neighbor_list = []
        for pos in neighbor_pos:
            # Skip identity
            if pos[0] == x and pos[1] == y:
                continue
                
            # Check if empty
            if not numpy.isnan(self.space[pos[0], pos[1]]):
                neighbor_list.append(int(self.space[pos[0], pos[1]]))
        
        return neighbor_list            
                   
    def get_village_position(self, village_id):
        """
        Get the position of a village based on its ID.
        """
        # Find the value that matches our ID in self.space, then reshape to a 2-element list.
        return numpy.reshape(numpy.where(m.space == village_id), (1, 2))[0].tolist()
        
    def get_tractor_position(self, tractor_id):
        """
        Get the position of a tractor based on its ID.
        """
        # Find the value that matches our ID in self.space, then reshape to a 2-element list.
        return numpy.reshape(numpy.where(m.tractorspace == tractor_id), (1, 2))[0].tolist()

    def get_tractor_neighbors(self, tractor_id, distance=1):
        """
        Get the position of a person based on their ID.
        """
        # Find the value that matches our ID in self.space, then reshape to a 2-element list.
        x, y = self.get_tractor_position(tractor_id)
        return self.get_neighbors(x, y, distance)   
            
    def move_tractor(self, tractor_id, x, y):
        """
        Move a tractor to a new (x, y) location as in function.
        """
        # Get original position
        original_position = self.get_tractor_position(tractor_id)
            
        # Check target location
        if not numpy.isnan(self.tractorspace[x, y]):
           raise ValueError("Unable to move tractor {0} to ({1}, {2}) since occupied.".format(tractor_id, x, y))
       
        # Otherwise, move by emptying and setting.
        self.tractorspace[original_position[0], original_position[1]] = numpy.nan
        self.tractorspace[x, y] = tractor_id
   
    def step_move(self):
        """
        Model step move function, which handles moves tractors randomly.
        """
        
        # Get a random order for the agents.
        random_order = range(self.num_tractors)
        numpy.random.shuffle(random_order)
        
        # Iterate in random order.
        for i in random_order:
            # Get current position
            x, y = self.get_tractor_position(i)

            neighbors = self.get_tractor_neighbors(i) # Get tractor neighbors

        """ 
        #BASE CASE:
            # Move our agent between -1, 0, +1 in each dimension (wraps)
            x_new = (x + numpy.random.randint(-1, 2)) % self.grid_size
            y_new = (y + numpy.random.randint(-1, 2)) % self.grid_size
            
            # Try to move them
            try:
                self.move_tractor(i, x_new, y_new)
            except ValueError:
                # Occupied, so fail.
                pass
        
        
        
        """
        #CASE 1
            # Move our agent to nearby village with most demand
                  
        if len(neighbors) > 0: #could be possible if that no neighbors
                to_go = neighbors[0] 
                for x in xrange(len(neighbors)-1):
                    if m.villages[neighbors[x+1]].total_silt_demand > m.villages[to_go].total_silt_demand:
                        to_go = neighbors[x+1]
                x1, y1 = self.get_village_position(to_go)
                x_new = x1
                y_new = y1
            
        if len(neighbors) < 1 or m.villages[to_go].total_silt_demand < 100: #if no neighbors move randomly as in base case (should be larger radius to find more demand)
                x_new = (x + numpy.random.randint(-1, 2)) % self.grid_size
                y_new = (y + numpy.random.randint(-1, 2)) % self.grid_size
            
                            
            # Try to move them
                try:
                    self.move_tractor(i, x_new, y_new)
                except ValueError:
                    # Occupied, so fail.
                    pass
             
        
        """
    
            # CASE 2
            # Move our agent to area of highest quality silt
                              
        if len(neighbors) > 0: #could be possible if that no neighbors
                to_go = neighbors[0] 
                for x in xrange(len(neighbors)-1):
                    if m.villages[neighbors[x+1]].silt_quality > m.villages[to_go].silt_quality:
                        to_go = neighbors[x+1]
                x1, y1 = self.get_village_position(to_go)
                x_new = (x1)
                y_new = (y1)
            
        if len(neighbors) < 1 or m.villages[to_go].total_silt_demand < 100: #if no neighbors move randomly as in base case (should be larger radius to find more demand)
                x_new = (x + numpy.random.randint(-1, 2)) % self.grid_size
                y_new = (y + numpy.random.randint(-1, 2)) % self.grid_size
           
                            
            # Try to move them
                try:
                    self.move_tractor(i, x_new, y_new)
                except ValueError:
                    # Occupied, so fail.
                    pass
         
         """    

    
    
    def step_interact(self):
        """ Interact: tractors apply silt on villages they pass by"""
        
        # Get a random order for the agents.
        random_order = range(self.num_tractors)
        numpy.random.shuffle(random_order)
        self.applied_this_step = 0 
        # Iterate in random order.
        for i in random_order:
            # Get neighbors
            neighbors = self.get_tractor_neighbors(i)
            
            # Iterate over neighbors
            for neighbor in neighbors:
                
                #apply neighboring villages that demand silt, not capital constrained, represents field manager not fleet
                
                self.villages[neighbor].total_silt_demand = (self.villages[neighbor].base_silt_demand + self.villages[neighbor].silt_quality * (self.villages[neighbor].silt_acres/self.villages[neighbor].cotton_acres) ) * (self.villages[neighbor].cotton_acres - self.villages[neighbor].silt_acres) 

                self.tractors[i].acres_applied = self.villages[neighbor].total_silt_demand
                self.villages[neighbor].silt_acres = self.villages[neighbor].silt_acres + self.villages[neighbor].total_silt_demand
                self.applied_this_step = self.applied_this_step + self.tractors[i].acres_applied 
        
        self.total_applied = self.applied_this_step + self.total_applied 
        
    def step(self):
        """Model step function"""
        self.step_interact()
        self.step_move()
        

        # Increment steps and track history.
        self.t += 1
        self.history_acres_applied.append(self.applied_this_step)
        self.history_total_acres_applied.append(self.total_applied)
    
```

#### Storing Data Code

```python
# Now, we'll define our methods to store a model's output

def store_model_parameters(model, run_output_path):
    """
    Store model parameters from a model to the run output path.
    """
    # Create parameters dictionary
    model_parameters = {"grid_size": model.grid_size,
                        "num_villages": model.num_villages,
                        "num_tractors": model.num_tractors,
                        }
    
    # Convert to dataframe and save
    model_parameters_df = pandas.DataFrame(model_parameters.items(),
                             columns=["parameter", "value"])
    model_parameters_df.to_csv(os.path.join(run_output_path, "parameters.csv"))
    

def store_model_csv(model, run_output_path):
    """
    Store CSV data from a model to the run output path.
    """
     
    # Create time series data frame
    tsdata_df = pandas.DataFrame(model.history_total_acres_applied,
                            columns=["total_silt_applied"])
    #tsdata_df["silt_applied"] = model.history_acres_applied
    #tsdata_df["Total_siltapplied"] = model.history_total_acres_applied
    
    # Save the dataframe
    tsdata_df.to_csv(os.path.join(run_output_path, "timeseries.csv"))

def store_model(model, output_path="output"):
    """
    Store a model to the model output path.
    """
    # First, we need to make sure the directory exists.
    try:
        os.makedirs(output_path)
    except:
        pass
    
    """
    Next, we need to create a unique timestamp for the model.
    We'll do that using a timestamp of the form: YYYYMMDD-Run#
    
    We then need to create that directory too.
    """
    timestamp_suffix = time.strftime("%Y%m%d")
    
    run_id = 0
    run_output_path = os.path.join(output_path,
                                 "run-{0}-{1}".format(timestamp_suffix,
                                                     run_id))
    # Get a unique 7 #
    while os.path.exists(run_output_path):
        run_id += 1
        run_output_path = os.path.join(output_path,
                                 "run-{0}-{1}".format(timestamp_suffix,
                                                     run_id))        

    try:
        os.makedirs(run_output_path)
    except:
        pass
    
    """
    Finally, we need to store data and figures to the path.
    """
    store_model_parameters(model, run_output_path)
    store_model_csv(model, run_output_path)
```
